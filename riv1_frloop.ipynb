# pritnt 'python' in reverse logic using for loop 5 times

```python
word = "python"

for j in range(5):              # Outer loop - 5 times
    for i in range(len(word)-1, -1, -1):   # Inner loop - reverse
        print(word[i], end="")
    print(" ",j)   # Next line
```

# i like python
# convert vowel to *
# * l*k* pyth*n

```python
text = 'i like python'
res = '' # empty string to store the output
for ch in text:# read each charecter one by one
    if ch in 'AEIOUaeiou': # check if charecter is ovwel
        res += '*' #replace the ovwel with star
    else: # if it is not a ovwel
        res += ch # keep the original charecter
        print(res)
```

| Character (`ch`) | Vowel? | `res` after update | Print?            |
| ---------------- | ------ | ------------------ | ----------------- |
| `i`              | ✅ Yes  | `*`                | ❌ No              |
| ` ` (space)      | ❌ No   | `* `               | ✅ `* `            |
| `l`              | ❌ No   | `* l`              | ✅ `* l`           |
| `i`              | ✅ Yes  | `* l*`             | ❌ No              |
| `k`              | ❌ No   | `* l*k`            | ✅ `* l*k`         |
| `e`              | ✅ Yes  | `* l*k*`           | ❌ No              |
| ` `              | ❌ No   | `* l*k* `          | ✅ `* l*k* `       |
| `p`              | ❌ No   | `* l*k* p`         | ✅ `* l*k* p`      |
| `y`              | ❌ No   | `* l*k* py`        | ✅ `* l*k* py`     |
| `t`              | ❌ No   | `* l*k* pyt`       | ✅ `* l*k* pyt`    |
| `h`              | ❌ No   | `* l*k* pyth`      | ✅ `* l*k* pyth`   |
| `o`              | ✅ Yes  | `* l*k* pyth*`     | ❌ No              |
| `n`              | ❌ No   | `* l*k* pyth*n`    | ✅ `* l*k* pyth*n` |


"The loop reads one character at a time from the string. If the character is a vowel, it appends * to res. Otherwise, it appends the original character and prints the current value of res. Since print(res) is inside the else block, output appears only when the current character is not a vowel."

```python
s = 'i like python'
res = ''
for ch in s:
    if ch in 'AEIOUaeiou':
        res += '*'
    else:
        res += ch
        print(res)
```

# for loop with break, continue, pass

```python
for i in range(2,10):
    if i == 7:
        break
    print(i)
        
```

```python
for i in range(2,10):
    if i == 7:
        continue
    print(i)
```

```python
for i in range(2,10):
    if i == 7:
        pass
    print(i)
```

# Nested for loop

```python
for i in range(1,10):
    for j in range(1,10):
        print(i,j)
```

"The outer loop controls the first number (i), and the inner loop controls the second number (j). For every single value of i, the inner loop runs completely from 1 to 9. Therefore, all pairs from (1,1) to (9,9) are printed, giving a total of 81 outputs."
