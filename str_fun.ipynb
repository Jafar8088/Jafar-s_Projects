# String Functions
### lower(), upper(), title(), capitalize()

```python
w = 'Welcome to Python'
len(w)
```

# Convert the entire given string to lower case

```python
w = "WElcOMe tO PytHON"
l = w.lower()
print(l)
```

# Convert the entire given string to upper case

```python
w = 'Welcome to Python'
U = w.upper()
print(U)
```

# Covert each first letter of the given string to upper case

```python
w = ' welcome to python'
t = w.title()
print(t)
```

# Convert the First to letter to upper case from the given string

```python
w = 'weLcome TO Python'
c = w.capitalize()
print(c)
```
