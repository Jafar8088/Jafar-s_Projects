# while Loops

## Syntax

###### Start
###### Condition
###### Increament / Decreament

```python
w=1 # start
while w<=10: # condition
    print(w, 'Welcome to Datamites')
    w=w+1 # increament or decreament
    print(w)
```

# Reverse Logic using while Loop

```python
a = 10
while a >= 1:
    print(a,'hello')
    a = a - 1
    print(a)
    
```

# Print the table using Loops

```python
num = int(input('Enter the table number : '))
i = 1
while i <= 10:
    print(num,'*',i,'=',num*i)
    i += 1
    print('Now value of i is Increamented by +1 : ',i)
```

# Print the tables in revers orders using while Loop


```python
num1 = int(input('Enter the Table number : '))
r = 10
while r >= 1:
    print(num1,'*',r,'=',num1*r)
    r = r-1
    print('Now value of r decreamented by -1 : ',r)
```
