# While Loop

## Syntax :

```python
#start value
#condition:
#increament / decreament
```

```python
i = 1 # strat value
while i <= 10: # condition
    print(i,'Welcome to python')
    i+=1 # increament / decreament
    print(i)
```

# Reverse Technique using while loop

```python
a = 10
while a >= 1:
    print(a,'hello')
    a-=1
    print(a)
```

# print the tables

```python
num = int(input('Enter the Table Nummber :'))
i = 1
while i <= 10:
    print(num,'*',i,'=',num * i)
    i += 1
```

# print the tables in reverse order

```python
num = int(input('Enter the Table Number :'))
i = 10
while i >= 1:
    print(num*i)
    i-=1
```
