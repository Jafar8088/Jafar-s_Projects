# Build Simple Calculator using if elif condition METHOD 1

```python
print('''
+ Addition
- Subtraction
* Multiplication
/ Division
''')
num1 = int(input('Enter the First Number : '))
num2 = int(input('Enter the Second Number : '))
opr = input('Enter the Operator : ')
if opr == '+':
    print(num1 + num2)
elif opr == '-':
    print(num1 - num2)
elif opr == '*':
    print(num1 * num2)
elif opr == '/':
    print(num1 / num2)
else:
    print('Invalid Operator')
```

# Build Simple Calculator using if condition METHOD 2

```python
x = int(input('Enter the First Number : '))
y = int(input('Enter the Second Number : '))
op = input('Enter the Operator : ')
if op == '+':
    print(x + y)
if op == '-':
    print(x - y)
if op == '*':
    print(x * y)
if op == '/':
    print(x / y)
if op != '+' and op != '-' and op != '*' and op != '/':
    print('Invalid Operator....!!!!')
        
```
