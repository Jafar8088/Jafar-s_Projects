# Indexing and Slicing

```python
# Take a string 'welcome to python' and print the letter 'e' by using +ve index

indx = 'welcome to python'
print(indx[1])
```

```python
#print letter 'n' by using -ve index

indx = 'welcome to python'
print(indx[-1])
```

```python
# print letter 't' by using +ve index and -ve index
indx = 'welcome to python'
print(indx[8])
print(indx[-4])
```

```python
#slicing 'welcome' from the above string by using +ve index

indx = 'welcome to python'
print(indx[0:7])
```

```python
indx[-18:-10]
```

```python
# slicing 'welcome' from the above string using -ve index

indx = 'welcome to python'
print(indx[-10:-18:-1])
```

```python
# print the entire string and increament by 2 using positive index and -ve index

indx = 'welcome to python'
print(indx[0::2])
print(indx[-18::2])
```

```python
#print the entire string by using -ve index and +ve index

indx = 'welcome to python'
print(indx[-18::])
print(indx[0::])
```

```python
# print the entire string in reverse order by using -ve index

indx = 'welcome to python'
print(indx[::-1]) #method 1
print(indx[-1:-18:-1]) # method 2
```

```python
#print 'to' from the above string by using +ve index
indx = 'welcome to python'
print(indx[8:10])
```

```python
# print 'ython' by using +ve index and -ve index
indx = 'welcome to python'
print(indx[12::])
print(indx[12:17])
print(indx[-5::])
```

```python
#slicing the 'welcome' and increament by 1 using +ve index 
indx = 'welcome to python'
print(indx[0:7:1])
```

```python
# slicing the 'welcome' and increament by 2
print(indx[0:7:2])
print(indx[-18:-10:2])
```

```python
#slicing 'wy' and icreament by 12 using +ve and -ve index
indx = 'welcome to python'
print(indx[0:13:12])
print(indx[-18:-4:12])
```

```python
#print 'emoc' by using -ve index
indx[-11:-15:-1]
```
