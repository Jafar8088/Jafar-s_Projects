# Operators

```python
# Arithmatic operators (+,-,*,/,%,**,//)
a = int(input('Enter the first number :'))
b = int(input('enter the second number :'))
add = a + b
sub = a - b
mul = a * b
div = a / b
mod = a % b
ex = a ** b
fd = a // b
print('Addition is :',add)
print('Subtraction is :',sub)
print('Multiplication is :',mul)
print('Division is :',div)
print('modulus is :',mod)
print('Exponents is :',ex)
print('floor division is :',fd)
```

```python
#Assignment Operators (=,+=,-=)
# x = 5
# x += 5 is same as x = x+5 means increment by 5

x = int(input('Enter the number'))
x += 5
print('increament value is :',x)
```

```python
# x -= 5 is same as x = x-3 means decrement by 5

x = int(input('enter the number'))
x -= 5
print('decreament value is :',x)
```

```python
# Comparision Operator (==,!=, >,<,>=,<=)
# Returns the values in 'True' or 'Flase'

a = int(input('Enter the Number :'))
b = int(input('Enter the number :'))
print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

```python
#Logical Operator 'and','or','not'
#'and'
a = int(input('enter the 1st number :'))
b = int(input('Enter the 2nd number :'))
print(a == 10 and a > b)
print(a == 10 and a == b)
```

```python
#'or'
a = int(input("Enter the first number :"))
b = int(input("Enter the second number :"))
print(a == 20 or a < b)
print(a == 20 or a != b)
```

```python
#not
a = int(input('enter the number :'))
print(a != 10)
```

```python
# Membership Operators 'in' and 'not in'
a = 'python'
'h' in a
```

```python
x = [10,20,30,40,50]
print(10 not in x)
print(60 not in x)
```

```python
#identity operator 'is' , 'not is'
a = 10
b = 10
print(a is b, a == b)
print(a is not b , a != b)
```

```python
#Bitwise Operator 'or','xor'
bin(10)
```

```python
x = 12
y = 13
print(bin(x))
print(bin(y))
print(x ^ y,bin(x ^ y))
```

```python
#Swapping the values of a and b
a = int(input("Enter the number for variable a :"))
b = int(input("Enter the number for variable b :"))

#formula for swapping the values of a and b

a = a + b
b = a - b
a = a - b
print("After swapping :",a)
print("After swapping :",b)
```
