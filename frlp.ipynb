# Implement range by using for loop

```python
for n in range(5):
    print(n)
```

```python
for n in range(1,5):
    print(n)
```

# Print 'welcome' 5 times using for loop

```python
for n in range(5):
    print(n,'welcome')
```

# Print the tables by using for loop

```python
for a in range(1,11):
    print(2,'*',a,'=',2*a)
```

# Reverse Logic in to print the tables using for loop

```python
# range(10,0,-1) #range(start_value,End_value_increament/decreament)
for l in range(10,0,-1):
    print(2,'*',l,'=',2*l)
```

# Implement using for loop

```python
for n in range(10,0,-1):
    print(n)
```

# print the number in reverse order and decreament by 2 using for loop

```python
for i in range(10,0,-2):
    print(i)
```

# print the given element given in the list in every line

```python
l = ['p',78,90.7,True]
for i in l: # for is used to loop the list
    print(i) # print 'i' value in each line
```

```python
l = ['p','y','t','H','o','n']
for i in l:
    print(i)
```

```python
#tuples
t = ('t',78,90.7,True)
for i in t:
    print(i)
```
